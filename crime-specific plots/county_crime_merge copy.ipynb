```python
import os
import pandas as pd
```

```python
def print_files_in_folder(folder_path):
    try:
        file_names = os.listdir(folder_path)
        return file_names
    except FileNotFoundError:
        print(f"Error: Folder '{folder_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

fns = print_files_in_folder("County_Crime")
```

```python
def clean_county(s):
    return s.replace("_", " ").replace(" County", "")

all_county_df = pd.read_csv("County_Crime/" + fns[0])

for fn in fns[1:]:
    if fn == ".ipynb_checkpoints":
        continue
    one_county_df = pd.read_csv("County_Crime/" + fn)
    all_county_df = pd.concat([all_county_df, one_county_df], ignore_index=True, axis=0)

all_county_df["County"] = all_county_df["county"].apply(clean_county)
all_county_df = all_county_df.drop("county", axis=1)
all_county_df = all_county_df.rename(columns={"year": "Year"})

all_county_df = all_county_df.sort_values(by=["Year", "County"])
all_county_df = all_county_df.set_index(["Year", "County"])

all_county_df = all_county_df.drop_duplicates() # remove the second set of 2014 rows

all_county_df
```

```python
all_county_df.to_csv("all_counties_crime.csv")
```
