This file is for converting frequencies of each crime to rates by dividig frequencies by county populations.

```python
import pandas as pd
```

```python
df = pd.read_csv("county_fips.csv") # retreived from https://github.com/kjhealy/fips-codes/blob/master/state_and_county_fips_master.csv
df_ca = df[df["state"] == "CA"]

def remove_county(s):
    if type(s) == str:
        return s.replace(" County", "")
    return s

df_ca["name"] = df_ca["name"].apply(remove_county)
fips_name_dict = df_ca.set_index("fips")["name"].to_dict()
```

```python
raw_df_80s = pd.read_csv("county_pops_1980s.csv")
```

```python
raw_df_80s = raw_df_80s.rename(columns={"Intercensal Estimates of the Resident Population of States and Counties": "fips"})

def in_CA(s):
    return s[:2] == "06" and s != "06000"

raw_df_80s["fips"] = raw_df_80s["fips"].astype(str)
raw_df_80s = raw_df_80s[raw_df_80s["fips"].apply(in_CA)]

def remove_comma_and_spaces(s):
    if type(s) == str:
        s = s.replace(",", "")
        return s.replace(" ", "")
    return s

raw_df_80s["Unnamed: 2"] = raw_df_80s["Unnamed: 2"].apply(remove_comma_and_spaces).astype(int)
raw_df_80s["Unnamed: 3"] = raw_df_80s["Unnamed: 3"].apply(remove_comma_and_spaces).astype(int)
raw_df_80s["Unnamed: 4"] = raw_df_80s["Unnamed: 4"].apply(remove_comma_and_spaces).astype(int)
raw_df_80s["Unnamed: 5"] = raw_df_80s["Unnamed: 5"].apply(remove_comma_and_spaces).astype(int)
raw_df_80s["Unnamed: 6"] = raw_df_80s["Unnamed: 6"].apply(remove_comma_and_spaces).astype(int)

raw_df_80s = raw_df_80s.reset_index()
```

```python
year_fips_pop_df_80s = pd.DataFrame(columns=["year", "fips", "Population"])
pop_cols = raw_df_80s.columns[3:]

for row_idx in range(raw_df_80s.shape[0]//2):
    fips = raw_df_80s["fips"].iloc[row_idx]
    extra_idx = 0
    for year in range(1980,1985):
        col_label = pop_cols[year-1980]
        pop = raw_df_80s.iloc[row_idx][col_label]
        year_fips_pop_df_80s.loc[5*row_idx + extra_idx] = [year, fips, pop]
        extra_idx += 1

for row_idx in range(raw_df_80s.shape[0]//2, raw_df_80s.shape[0]):
    fips = raw_df_80s["fips"].iloc[row_idx]
    extra_idx = 0
    for year in range(1985,1990):
        col_label = pop_cols[year-1985]
        pop = raw_df_80s.iloc[row_idx][col_label]
        year_fips_pop_df_80s.loc[5*row_idx + extra_idx] = [year, fips, pop]
        extra_idx += 1
```

```python
def fips_to_name(f):
    return fips_name_dict[int(f)]

year_fips_pop_df_80s["County"] = year_fips_pop_df_80s["fips"].apply(fips_to_name)
year_pop_county_df_80s = year_fips_pop_df_80s.drop(columns=["fips"])
year_pop_county_df_80s
```

```python
raw_df_90s = pd.read_csv('county_pops_1990s.txt', sep='\t') # For tab-separated
```

```python
year_fips_pop_df_90s = pd.DataFrame(columns=["year", "fips", "Population"])
for row_idx in range(13, raw_df_90s.shape[0]):
    strs_list = raw_df_90s.iloc[row_idx][0].split(" ")
    state_fips = strs_list[1][:2]
    if state_fips == "06": # ensuring fips is from California
        year, fips = strs_list[0], strs_list[1]
        total_pop = 0
        for s in strs_list[2:]:
            if len(s) != 0:
                total_pop += int(s)
        year_fips_pop_df_90s.loc[row_idx-13] = [year, fips, total_pop]
```

```python
year_fips_pop_df_90s = year_fips_pop_df_90s.reset_index()
year_fips_pop_df_90s = year_fips_pop_df_90s.drop(columns=["index"])
year_fips_pop_df_90s["year"] = year_fips_pop_df_90s["year"].astype(int)
year_fips_pop_df_90s
```

```python
def fips_to_name(f):
    return fips_name_dict[int(f)]

year_fips_pop_df_90s["County"] = year_fips_pop_df_90s["fips"].apply(fips_to_name)
year_pop_county_df_90s = year_fips_pop_df_90s.drop(columns=["fips"])
year_pop_county_df_90s
```

```python
raw_df_00s = pd.read_csv("county_pops_2000s.csv")
raw_df_00s = raw_df_00s.iloc[2:62]
raw_df_00s = raw_df_00s.rename(columns={"table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)": "County"})
raw_df_00s
```

```python
def remove_dots_and_County(s):
    if type(s) == str:
        s = s.replace(" County", "")
        return s.replace(".", "")
    return s

raw_df_00s["County"] = raw_df_00s["County"].apply(remove_dots_and_County).astype(str)

raw_df_00s.columns = list(raw_df_00s.iloc[0])
raw_df_00s = raw_df_00s.iloc[2:]
raw_df_00s = raw_df_00s.rename(columns={"nan": "County"})
raw_df_00s.head()
```

```python
year_pop_county_df_00s = pd.DataFrame(columns=["year", "Population", "County"])

counties = list(raw_df_00s["County"])

def remove_comma(s):
    if type(s) == str:
        return s.replace(",", "")
    return s

row_idx = 0
for county in counties:
    for year in range(2000, 2010):
        pop = int(remove_comma(raw_df_00s[raw_df_00s["County"] == county][str(year)].iloc[0]))
        year_pop_county_df_00s.loc[row_idx] = [year, pop, county]
        row_idx += 11

year_pop_county_df_00s = year_pop_county_df_00s.reset_index()
year_pop_county_df_00s = year_pop_county_df_00s.drop("index", axis=1)
year_pop_county_df_00s
```

```python
year_pop_county_df_00s
```

```python
raw_df_10s = pd.read_csv("county_pops_2010s.csv")
raw_df_10s = raw_df_10s.iloc[2:62]
raw_df_10s = raw_df_10s.rename(columns={"table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)": "County"})
raw_df_10s
```

```python
def remove_dots_and_CC(s):
    if type(s) == str:
        s = s.replace(" County, California", "")
        return s.replace(".", "")
    return s

raw_df_10s["County"] = raw_df_10s["County"].apply(remove_dots_and_CC).astype(str)

raw_df_10s.columns = list(raw_df_10s.iloc[0])
raw_df_10s = raw_df_10s.iloc[2:]
raw_df_10s = raw_df_10s.rename(columns={"nan": "County"})
raw_df_10s.head()
```

```python
year_pop_county_df_10s = pd.DataFrame(columns=["year", "Population", "County"])

counties = list(raw_df_10s["County"])

row_idx = 0
for county in counties:
    for year in range(2010, 2020):
        pop = int(remove_comma(raw_df_10s[raw_df_10s["County"] == county][str(year)].iloc[0]))
        year_pop_county_df_10s.loc[row_idx] = [year, pop, county]
        row_idx += 11

year_pop_county_df_10s = year_pop_county_df_10s.reset_index()
year_pop_county_df_10s = year_pop_county_df_10s.drop("index", axis=1)
```

```python
raw_df_20s = pd.read_csv("county_pops_2020s.csv")
```

```python
raw_df_20s = raw_df_20s.drop("Unnamed: 1", axis=1)
raw_df_20s = raw_df_20s.iloc[2:62]
raw_df_20s = raw_df_20s.rename(columns={"table with row headers in column A and column headers in rows 3 through 4 (leading dots indicate sub-parts)": "County"})

raw_df_20s["County"] = raw_df_20s["County"].apply(remove_dots_and_CC).astype(str)

raw_df_20s

raw_df_20s.columns = list(raw_df_20s.iloc[0])
raw_df_20s = raw_df_20s.iloc[2:]
raw_df_20s = raw_df_20s.rename(columns={"nan": "County"})
```

```python
year_pop_county_df_20s = pd.DataFrame(columns=["year", "Population", "County"])

row_idx = 0
for county in counties:
    for year in range(2020, 2025):
        pop = int(remove_comma(raw_df_20s[raw_df_20s["County"] == county][str(year)].iloc[0]))
        year_pop_county_df_20s.loc[row_idx] = [year, pop, county]
        row_idx += 6

year_pop_county_df_20s = year_pop_county_df_20s.reset_index()
year_pop_county_df_20s = year_pop_county_df_20s.drop("index", axis=1)
```

```python
acc_df = pd.read_csv("all_counties_crime.csv") # goal is to make a column with county population for each year

### Turn all crime freqs to rates
crime_types = acc_df.columns[2:]

def remove_comma(s):
    if type(s) == str:
        return s.replace(",", "")
    return s

for ct in crime_types:
    acc_df[ct] = acc_df[ct].apply(remove_comma)
    acc_df[ct] = acc_df[ct].astype(float)

ca_agg = acc_df.groupby(["Year"]).sum()
ca_agg["County"] = pd.Series(ca_agg.shape[0] * ["California"], index=range(1985, 2024))
ca_agg = ca_agg.reset_index()

acc_df = pd.concat([ca_agg, acc_df])
```

```python
year_pop_county_df_all_years = pd.concat([year_pop_county_df_80s, year_pop_county_df_90s, year_pop_county_df_00s, year_pop_county_df_10s, year_pop_county_df_20s], ignore_index=True).sort_values(["County", "year"])

ca_agg_pop = year_pop_county_df_all_years.groupby(["year"]).sum()
ca_agg_pop["County"] = pd.Series(ca_agg_pop.shape[0] * ["California"], index=range(1980, 2025))
ca_agg_pop = ca_agg_pop.reset_index()

year_pop_county_df_all_years = pd.concat([ca_agg_pop, year_pop_county_df_all_years])
```

```python
acc_w_pop_df = pd.merge(left=acc_df, right=year_pop_county_df_all_years, left_on=["Year", "County"], right_on=["year", "County"])
acc_w_pop_df = acc_w_pop_df.drop(columns=["year"])

for ct in crime_types:
    acc_w_pop_df[ct] = acc_w_pop_df[ct].apply(remove_comma)
    acc_w_pop_df[ct] = acc_w_pop_df[ct].astype(float)

acc_w_pop_df[list(crime_types)] = acc_w_pop_df[list(crime_types)].div(acc_w_pop_df["Population"], axis=0)
print(acc_w_pop_df.shape)
acc_w_pop_df.head()
```

```python
acc_w_pop_df.to_csv("all_counties_crime_w_pop.csv")
```
